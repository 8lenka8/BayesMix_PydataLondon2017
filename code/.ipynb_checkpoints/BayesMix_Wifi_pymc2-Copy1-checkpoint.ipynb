```python
# A helper function to automatically colour code some continuous numeric or ordinal values 

#examle colour = 'rgb(0,0,255)'
def rgb(minimum, maximum, value):
    minimum, maximum = float(minimum), float(maximum)
    ratio = 2 * (value-minimum) / (maximum - minimum)
    b = int(max(0, 255*(1 - ratio)))
    r = int(max(0, 255*(ratio - 1)))
    g = 255 - b - r
    return r, g, b
```

# Which algorithm to choose?



What kind of algorithm can help us disentangle the different wifi points that give raise to this data?...

* I need an unsupervided learning method (cluster analysis) that permits each data point to potentially belong to more than one cluster
* I have an idea about how my data is distributed - from theory or observation
* Ideally I also want to calculate it's probability of belonging to each cluster


Some of the options are:

* Fuzzy c-means cluster
* LDA
* BMM

Can BMM may help us disentangle the different wifi points that give raise to this data?...


It looks like BMM is promising - because I believe my data is actually Gaussian (Normally distributed), there is not a lot of it and I don't know enough about other parameters and I want to capture my uncertainty about the parameters

```python
data = wifi_df[['latitude','longitude']].values

nvars =  data.shape[1]  # number of dimensions
nobs = data.shape[0]    # number of observations
print("nobs: ", nobs)
print("nvars: ", nvars) 
```

### Hierarchical cluster to set number of clusters and cluster centres priors
-- diagram of graphical model and corresponding intuition: 3 normals with mu and std, 1/3 prior and stick breaking data cluster membership

```python
def hierarchical_cl(df, plot_tree=True, maxDistance = 0.0005):
    
    # generate the hirarchical cluster tree based on eukledian distance matrix between pairs of coordinates
    distanceM = linkage(df.values, method='ward', metric='euclidean')        
          
    # plot dendrogram    
    if plot_tree == True:
        plt.figure(figsize=(25, 10))
        plt.title('Hierarchical Clustering Dendrogram')
        plt.xlabel('reading')
        plt.ylabel('distance')
        dendrogram(
            distanceM,
            truncate_mode='lastp',
            show_leaf_counts=True,
        )
        plt.show()
    

    #get cluster ids back
    clusters = fcluster(distanceM, maxDistance, criterion='distance')
    
    # optimal number of clusters
    cluster_counts = np.unique(clusters, return_counts=True) 
    cluster_counts = [x for x in zip(cluster_counts[0],cluster_counts[1])]
    nclusters_prior = len(cluster_counts) 
    
    # median long and lat within each cluster
    df['cl'] = clusters
    
    med = df.groupby('cl')    
    cl_medians_df = pd.DataFrame(columns=[u'latitude', u'longitude'])
    
    for grp, data in enumerate(med):

        (BSSID, coords) = data
            
        median = coords[['latitude', 'longitude']].median()

        cl_medians_df.loc[BSSID] = median
    
    cl_medians_prior = cl_medians_df.values
    
    return nclusters_prior, cl_medians_prior
```

### Using fixed number of clusters and cluster centres estimated by hierarchical cluster algorithm

```python
nclusters, cl_medians = hierarchical_cl(df = wifi_df[['latitude','longitude']])
print("nclusters: ", nclusters) 
```

```python
print("Actual number of BSSIDs is: {}".format(len(BSSID_df)))
```

```python
# If we don't know the number of clusters, we can estimate it using Truncated Poisson distribution 
# with parameter mu = our best guess about the rate of wifi points per sq area
# parameter k = 0 minimum truncation point i.e. minimum 1 cluster, number of clusters can't be negative 
# assume one wifi point's location is independent of another - they are not really, but will do for now
#nclusters_prior, cl_medians = = hierarchical_cl(df = wifi_df[['latitude','longitude']])
#print("nclusters_prior: ", nclusters_prior)
#nclusters = pm.TruncatedPoisson('nclusters', mu = nclusters_prior, k=0)             
#print("nclusters: ", nclusters.value)
```

```python
# use 'stick-breaking' Dirichlet process to randomly generate the cluster probabilities that all add up to 1
# initialise non-informative priors for probabilities (theta=) with non-zero starting values (value=)
theta_init = np.ones(nclusters)/ (nclusters) 
cluster_probs = pm.Dirichlet('cluster_probs', theta = theta_init, value = theta_init[1:], observed=False)
print("cluster_probs: ", cluster_probs.value)
```

```python
# assign each data point to it's cluster according to the cluster probabilities generated by Dirichlet process
data_cluster_id = pm.Categorical('data_cluster_id', p = cluster_probs, size = nobs) 
print("data_cluster_id: ", data_cluster_id.value[0:10])
```

```python
print("data_cluster_id: ", np.unique(data_cluster_id.value, return_counts=True))
```

```python
# Model cluster centres
# I want to forse std_dev of the mean to be small because I know typical wifi signal range 
# and the range of values it's mean can take is even narrower (more precice) - CLT
std_dev = 0.001
tau = std_dev**-2  # Normal in pymc2 is defined in terms of precision rather than standard deviation

cluster_means = []

for j in range(nclusters):
    if j<2:
        print(cl_medians[j])
    mu = pm.Normal('posterior_cluster_mean%d' % j, cl_medians[j] , tau, size=nvars) # initiate mus at some of static medians
    if j<2:
        print(mu.value)
    cluster_means.append(mu)

cluster_means = np.array(cluster_means, object) # Combine means arrays into matrix
```

```python
# Model cluster 'precision' (spread and tilt)
tau_init = np.eye(nvars) # uniformative wide priors for precision of covariance matrix
nu = nvars+1 # prior for degrees of freedom (should be > number of variables)

cluster_precision = []

for j in range(nclusters):
    cov = pm.Wishart('cov_matrix_inv%d' % j, n=4, Tau=tau_init)
    cluster_precision.append(cov)
    
cluster_precision = np.array(cluster_precision, object)

# Could force covariance = 0 as we expect round wifi signal range...
# ... but best not to for now, as the signal is reflected of building walls and so can be oblong 
```

```python
####### model
    
@pm.deterministic
def mean(data_cluster_id = data_cluster_id, cluster_means = cluster_means):
    return cluster_means[data_cluster_id]

@pm.deterministic
def prec(data_cluster_id = data_cluster_id, cluster_precision = cluster_precision):
    return cluster_precision[data_cluster_id]
    
@pm.observed
def obs(value=data, mean=mean, prec=prec):
    # multivariate normal log-likelihood
    return sum(pm.mv_normal_like(v, m, T) for v,m,T in zip(value, mean, prec))
```

```python
model = pm.Model([cluster_probs, data_cluster_id, obs, prec, mean])
graph = pm.graph.graph(model)
```

```python
##### mcmc sampler - expect to take hours or even overnight before converges in some applications 

mcmc = pm.MCMC(model)
mcmc.sample(iter=600000,  thin=500, burn=500)
```

```python
# estimated cluster centres
a_lat = []
a_long = []  
for j in range(len(cluster_means)):
    a_lat.append(mcmc.trace("posterior_cluster_mean%d" % j)[:][:,0].mean())
    a_long.append(mcmc.trace("posterior_cluster_mean%d" % j)[:][:,1].mean())

cl_centres = [[a,b] for a,b in zip(a_lat,a_long)]
```

```python
# estimated covariance matrices
## invert cluster precision tau back into standard deviation
cl_cov = []
for j in range(nclusters):
    tau_j = mcmc.trace('cov_matrix_inv%d' % j)[:47].mean(axis=0) 
    #stdev.append(math.sqrt(std_j**-1))
    std_j = np.linalg.inv(tau_j)
    cl_cov.append(std_j)
    if j==0:
        print (j)
        print("precision: ")
        print(tau_j)
        print(" covariance:")
        print(std_j)
 
        # Checks: Semi-positive definite if all eigenvalues are >=0 or if there exists a Cholesky decomposition 
        print ("eigvals")
        print (np.linalg.eigvals(cl_cov[j]))
        print ("cholesky")
        print (np.linalg.cholesky(cl_cov[j]))
```

```python
# estimated cluster probabilities
cluster_probs_avg = mcmc.trace("cluster_probs")[:].mean(axis=0)
cluster_probs_avg = np.append(cluster_probs_avg, (1-sum(cluster_probs_avg)))
cluster_probs_avg[0:nclusters]
```

```python
# create matrix of probabilities of each point belonging to each cluster with width = nobs length = nclusters
pr = []
for j in range(nclusters):
    probs = multivariate_normal.pdf(data.tolist(), mean = cl_centres[j], cov = cl_cov[j].tolist())
    pr.append(probs)

# transpose pr list to be width = nclusters and length = nobs
probabilities = np.asarray(pr).T
probs_max = [ [probabilities[i].argmax(),probabilities[i].max()] for i in range(nobs) ]
```

```python
def cov2corr(cov_matrix):
    corr = cov_matrix[0][1]/(np.sqrt(cov_matrix[0][0])*np.sqrt(cov_matrix[1][1]))
    return corr
    
def corr2angle(corr):
    angle = np.arctan(corr)
    return angle


def colours(d):  
    
    minidx = min(d)
    maxidx = max(d)
        
    #colour = 'rgb(0,0,255)'
    dff = pd.DataFrame(d, columns= ['data_cluster_id'])
    dff['colour'] = dff['data_cluster_id'].map(lambda x: 'rgb({},{},{})'.format(rgb(minidx, maxidx, x)[0] , rgb(minidx, maxidx, x)[1] , rgb(minidx, maxidx, x)[2]))
    colours = dff['colour']
    
    return colours
```

```python
from jinja2 import Template

from folium.element import JavascriptLink, Figure #,  MacroElement
from folium.map import Popup, Icon, Marker


class ellipse(Marker):
    def __init__(self, location):
        """Creates a Ellipse marker plugin to append into a map like CircleMarker
        Parameters
        ----------
        location: list of 3 lists, default None
            1) Latitude and Longitude of centre (in my case centre of cluster of data points)
            2) Radius 1, Radius 2 (in my case standard deviation of latitude and of longitude of data points)
            3) Angle of tilt (in my case correlation between longitude and latitude obserations)
            Example:
            [[51.5, -0.09], [500, 100], 90]
        
        """
        super(ellipse, self).__init__(location)
        self._name = 'ellipse'

        #var ellipse = L.ellipse([51.5, -0.09], [500, 100], 90).addTo(map);
        self._template = Template(u"""
            {% macro script(this, kwargs) %}
            
            var {{this.get_name()}} = L.ellipse(
                [{{this.location[0][0]}},{{this.location[0][01]}}],
                [{{this.location[1][0]}},{{this.location[1][1]}}],
                {{this.location[2]}}
                )
                .addTo({{this._parent.get_name()}});
            {% endmacro %}
            """)
                
    def render(self, **kwargs):
        super(ellipse, self).render(**kwargs)

        figure = self.get_root()
        assert isinstance(figure, Figure), ("You cannot render this Element "
                                            "if it's not in a Figure.")

        figure.header.add_children(
            #JavascriptLink("https://cdnjs.cloudflare.com/ajax/libs/leaflet.markercluster/0.4.0/leaflet.markercluster.js"),  # noqa
            #JavascriptLink("https://raw.githubusercontent.com/jdfergason/Leaflet.Ellipse/master/l.ellipse.min.js"),
            JavascriptLink("http://d8a.solutions/elenas_cdn/l.ellipse.js"),    
        name='ellipse')
```

### Visualise resulting clusters using estimated means and variance-covariance marices

```python
wifiMapTry  = folium.Map(location= map_center, zoom_start=19)

for router in BSSID_df[['BSSID']].values.tolist():
    
    router_df = newdf[newdf['BSSID'] == router[0]][['latitude','longitude']]
    colours = newdf[newdf['BSSID'] == router[0]]['colour']
    
    for record in router_df.index:
        folium.CircleMarker(router_df.loc[record], fill_color=colours.loc[record], radius=3, fill_opacity=0.5).add_to(wifiMapTry)


# standard deviation ellipses
for cl in range(nclusters):
    var, U = np.linalg.eig(cl_cov[cl])
    angle_xy = 180. / np.pi * np.arccos(np.abs(U[0, 0]))
    std_x = 10 * np.sqrt(5.991 * var[0]) # 'radius' in metres sideways is proportional to standard deviation of latitude
    std_y = 10 * np.sqrt(5.991 * var[1])
    #0.00001° ≈ 1m
    location = [cl_centres[cl], [std_x, std_y], -angle_xy-90]
    ellipse(location).add_to(wifiMapTry)
    
    
# cluster centres estimated by mcmc and cluster centre priors (from hierarchical cl)
for cl in range(nclusters):
    folium.CircleMarker(cl_centres[cl]
                            , fill_color='white'
                            , radius=3
                            , fill_opacity=1
                            , popup='{} {} {}'.format(cl,cluster_probs_avg[cl],cl_centres[cl])
                           ).add_to(wifiMapTry)

```
