```python
# A helper function to automatically colour code some continuous numeric or ordinal values 

#examle colour = 'rgb(0,0,255)'
def rgb(minimum, maximum, value):
    minimum, maximum = float(minimum), float(maximum)
    ratio = 2 * (value-minimum) / (maximum - minimum)
    b = int(max(0, 255*(1 - ratio)))
    r = int(max(0, 255*(ratio - 1)))
    g = 255 - b - r
    return r, g, b
```

### Hierarchical cluster to set number of clusters and cluster centres priors 

```python
-- diagram of graphical model and corresponding intuition: 
    3 normals with mu and std,
    1/3 prior and stick breaking
    data cluster membership 
```

```python
def hierarchical_cl(df, plot_tree=True, maxDistance = 0.0001):
    
    # generate the hirarchical cluster tree based on eukledian distance matrix between pairs of coordinates
    distanceM = linkage(df.values, method='ward', metric='euclidean')        
          
    # plot dendrogram    
    if plot_tree == True:
        plt.figure(figsize=(25, 10))
        plt.title('Hierarchical Clustering Dendrogram')
        plt.xlabel('reading')
        plt.ylabel('distance')
        dendrogram(
            distanceM,
            truncate_mode='lastp',
            show_leaf_counts=True,
        )
        plt.show()
    

    #get cluster ids back
    clusters = fcluster(distanceM, maxDistance, criterion='distance')
    
    # optimal number of clusters
    cluster_counts = np.unique(clusters, return_counts=True) 
    cluster_counts = [x for x in zip(cluster_counts[0],cluster_counts[1])]
    nclusters_prior = len(cluster_counts) 
    
    # median long and lat within each cluster
    df['cl'] = clusters
    
    med = df.groupby('cl')    
    cl_medians_df = pd.DataFrame(columns=[u'latitude', u'longitude'])
    
    for grp, data in enumerate(med):

        (BSSID, coords) = data
            
        median = coords[['latitude', 'longitude']].median()

        cl_medians_df.loc[BSSID] = median
    
    cl_medians_prior = cl_medians_df.values
    
    return nclusters_prior, cl_medians_prior
```

```python
data = wifi_df[['latitude','longitude']].values

nvars =  data.shape[1]  # number of dimensions
nobs = data.shape[0]    # number of observations
print("nobs: ", nobs)
print("nvars: ", nvars)  
```

### Using fixed number of clusters and cluster centres estimated by hierarchical cluster algorithm

```python
nclusters, cl_medians = hierarchical_cl(df = wifi_df[['latitude','longitude']])
print("nclusters: ", nclusters) 
```

```python
# Alternatively if we don't know the number of clusters, we can estimate it using Truncated Poisson distribution 
# with parameter mu = our best guess about the rate of wifi points per sq area
# parameter k = 0 minimum truncation point i.e. minimum 1 cluster, number of clusters can't be negative 
# assume one wifi point's location is independent of another - they are not really, but will do for now
#nclusters_prior, cl_medians = = hierarchical_cl(df = wifi_df[['latitude','longitude']])
#print("nclusters_prior: ", nclusters_prior)
#nclusters = pm.TruncatedPoisson('nclusters', mu = nclusters_prior, k=0)             
#print("nclusters: ", nclusters.value)
```

```python
# use 'stick-breaking' Dirichlet process to randomly generate the cluster probabilities that all add up to 1
# initialise non-informative priors for probabilities (theta=) with non-zero starting values (value=)
theta_init = np.ones(nclusters)/ (nclusters) 
cluster_probs = pm.Dirichlet('cluster_probs', theta = theta_init, value = theta_init[1:], observed=False)
print("cluster_probs: ", cluster_probs.value)
```

```python
# assign each data point to it's cluster according to the cluster probabilities generated by Dirichlet process
data_cluster_id = pm.Categorical('data_cluster_id', p = cluster_probs, size = nobs) 
print("data_cluster_id: ", data_cluster_id.value[0:10])
```

```python
print("data_cluster_id: ", np.unique(data_cluster_id.value, return_counts=True))
```

```python
# Model cluster centres
# I want to forse std_dev of the mean to be small because I know typical wifi signal range 
# and the range of values it's mean can take is even narrower (more precice) - CLT
std_dev = 0.001
tau = std_dev**-2  # Normal in pymc2 is defined in terms of precision rather than standard deviation

cluster_means = []

for j in range(nclusters):
    if j<2:
        print(cl_medians[j])
    mu = pm.Normal('posterior_cluster_mean%d' % j, cl_medians[j] , tau, size=nvars) # initiate mus at some of static medians
    if j<2:
        print(mu.value)
    cluster_means.append(mu)

cluster_means = np.array(cluster_means, object) # Combine means arrays into matrix
```

```python
# Model cluster 'precision' (spread and tilt)
tau_init = np.eye(nvars) # uniformative wide priors for precision of covariance matrix
nu = nvars+1 # prior for degrees of freedom (should be > number of variables)

cluster_precision = []

for j in range(nclusters):
    cov = pm.Wishart('cov_matrix_inv%d' % j, n=4, Tau=tau_init)
    cluster_precision.append(cov)
    
cluster_precision = np.array(cluster_precision, object)

# Could force covariance = 0 as we expect round wifi signal range...
# ... but best not to for now, as the signal is reflected of building walls and so can be oblong 
```

```python
####### model
    
@pm.deterministic
def mean(data_cluster_id = data_cluster_id, cluster_means = cluster_means):
    return cluster_means[data_cluster_id]

@pm.deterministic
def prec(data_cluster_id = data_cluster_id, cluster_precision = cluster_precision):
    return cluster_precision[data_cluster_id]
    
@pm.observed
def obs(value=data, mean=mean, prec=prec):
    # multivariate normal log-likelihood
    return sum(pm.mv_normal_like(v, m, T) for v,m,T in zip(value, mean, prec))
```

```python
model = pm.Model([cluster_probs, data_cluster_id, obs, prec, mean])
graph = pm.graph.graph(model)
```

```python
##### mcmc sampler - expect to take hours or even overnight before converges in some applications 

mcmc = pm.MCMC(model)
mcmc.sample(iter=500000,  thin=1000, burn=1000)
```
